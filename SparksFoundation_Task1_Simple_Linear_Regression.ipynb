### Extract the dependent and independent variable from the dataset

```python
x = df['Hours'].values
y = df['Scores'].values
x,y
```

## Train Test split Data

```python
from sklearn.model_selection import train_test_split
xtrain , xtest , ytrain , ytest = train_test_split(x,y,random_state=0,test_size=0.2)
xtrain_reshape = xtrain.reshape(-1,1)
ytrain_reshape = ytrain.reshape(-1,1)
xtest_reshape = xtest.reshape(-1,1)
ytest_reshape = ytest.reshape(-1,1)
xtrain_reshape
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain_reshape,ytrain_reshape)
```

```python
y_pred = model.predict(ytrain_reshape)
y_pred
```

```python
e = float(input("Enter any number:"))
e_reshaped = np.array(e).reshape(1, -1)
predict = model.predict(e_reshaped)
print(predict)
```

### Data Visualization

In the below graph we have plotted the graph using a data visualization library which is matplotlib library and able to predict by the model

```python
plt.scatter(x,y,color="red")
plt.plot(xtest_reshape,model.predict(xtest_reshape),color="blue")
plt.title("Hours Vs Scores")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```
